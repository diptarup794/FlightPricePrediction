```python
df.drop(columns=['Unnamed: 0'], inplace=True)
df.head()
```

```python
# Convert the 'Dep_Date' column to datetime format
df['Dep_Date'] = pd.to_datetime(df['Dep_Date'], format='%d/%m/%y')

# Extract Day, Month, and Year and create new columns
df['Day'] = df['Dep_Date'].dt.day
df['Month'] = df['Dep_Date'].dt.month
df['Year'] = df['Dep_Date'].dt.year

df.head()
```

```python
# Split 'Dep_time' into Hour and Minute
df['Hour'], df['Minute'] = zip(*df['Dep_time'].map(lambda x: x.split(':')))
df['Hour'] = df['Hour'].astype(int)
df['Minute'] = df['Minute'].astype(int)
# Show the DataFrame
df.head()
```

```python
# Create a 'Day_of_Week' column
df['Day_of_Week'] = df['Dep_Date'].dt.day_name()
df.head(100)
```

```python
# Columns to be one-hot encoded
columns_to_encode = [' Origin', ' Destination', 'Cabin', 'Booking Class', 'Airline', 'Flight number', 'Day_of_Week']
df[columns_to_encode]
```

```python
df = pd.get_dummies(df, columns=columns_to_encode)
df.head()
```

```python
booking_cols = [col for col in df.columns if 'num of bookings before:' in col]

print(len(booking_cols))
df[booking_cols] = df[booking_cols].fillna(-1)
#
df[booking_cols].head()
# print("----")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


# Separate features and target variable
X = df.drop(columns=['Booking Class Fare USD'])
y = df['Booking Class Fare USD']

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Display the shape of the training and test sets to confirm
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
from math import sqrt

# Initialize the XGBoost Regressor model
xgb_model = XGBRegressor(objective='reg:squarederror', n_estimators=100, random_state=42)

# Train the model on the training data
xgb_model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = xgb_model.predict(X_test)

print(len(y_pred))
print(len(y_test))

# Evaluate the model
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = sqrt(mse)

mae, rmse
```

```python
import pandas as pd


df_metrics = pd.DataFrame({'y_test':y_test,'y_pred':y_pred})

# Calculate the actual vs predicted difference
df_metrics['actual_vs_prd'] = df_metrics['y_test'] - df_metrics['y_pred']

# Print the DataFrame
df_metrics
```

```python
from sklearn.metrics import mean_squared_error, r2_score, explained_variance_score
import numpy as np

# Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")

# Mean Absolute Percentage Error
mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
print(f"Mean Absolute Percentage Error: {mape}")

# R-Squared
r2 = r2_score(y_test, y_pred)
print(f"R-Squared: {r2}")

# Explained Variance Score
evs = explained_variance_score(y_test, y_pred)
print(f"Explained Variance Score: {evs}")

# Median Absolute Error
medae = np.median(np.abs(y_test - y_pred))
print(f"Median Absolute Error: {medae}")
```
